```python
df = pd.read_csv('BloodPressure.csv')
df
```

```python
x_AgeWeight = df[['Age','Weight_pounds']]
y = df.Blood_pressure.values.reshape(-1,1)
```

```python
model = LinearRegression()
model.fit(x_AgeWeight,y)
print(model.coef_)
print(model.intercept_)
```

```python
model.score(x_AgeWeight,y)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
y_predict1=model.predict(x_AgeWeight)
print('MAE =', mean_absolute_error(y,y_predict1))
print('MSE =', mean_squared_error(y,y_predict1))
```
